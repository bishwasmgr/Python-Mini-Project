```python
import random
print("Welcome to the Number Guessing Game!")
print("Choose a difficulty level:")
print("1 - Easy (1 to 20)")
print("2 - Medium (1 to 50)")
print("3 - Hard (1 to 100)")
difficulty_level = input("Enter 1, 2, or 3: ")

#validate input properly
while difficulty_level not in ["1", "2", "3"]:
    print("Invalid choice. Please enter only 1, 2, or 3.")
    difficulty_level = input("Enter 1, 2, or 3: ")

#Changing difficulty_level into a range:
if difficulty_level == "1":
    max_value = 20
elif difficulty_level == "2":
    max_value = 50
else: 
    max_value = 100

secret_number = random.randint(1, max_value)
print(max_value)
guess = int(input(f"Guess a number between 1 and {max_value}: "))
count = 1
while guess != secret_number:
    count +=1
    if guess > secret_number:
        print("Too high! Try again.")
    else:
        print("Too low! Try again.")
    guess = int(input("Guess again: "))
print(f"Congratulations! You guessed it! on {count} try")
```
