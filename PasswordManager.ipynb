```python
def load_passwords():
    passwords = []
    try: 
        with open("passwords.csv", "r") as file:
            next(file)
            for line in file:
                parts = line.strip().split(",")
                record = {
                            "website": parts[0],
                            "username": parts[1],
                            "password": parts[2]
                        }
                passwords.append(record)
                 
    except Exception:
        pass
    return passwords                
```

```python
def save_passwords(passwords):
    with open("passwords.csv", "w") as file:
        file.write("website,username,password\n")
        for record in passwords:
            file.write(record["website"] + "," + record["username"] + "," + record["password"] + "\n")
    
```

```python
def add_password(passwords):
    website = input("Enter the Name of Website: ")
    username = input("Enter the Name of Username: ")
    password = input("Enter the Name of Password: ")
    record = {
        "website": website,
        "username": username,
        "password": password
    }
    print("Password added successfully!")
    passwords.append(record)
    save_passwords(passwords)
```

```python
def search_password(passwords):
    website = input("Enter the Name of Website you want to find password: ")
    found = False
    for record in passwords:
        if record["website"].lower()  == website.lower():
            print(record)
            found = True
    if not found:
        print("No Record Found")          
```

```python
def main():
    passwords = load_passwords()

    while True:
        print("\n====== PASSWORD MANAGER ======")
        print("1. View all accounts")
        print("2. Add a new password")
        print("3. Search for a password")
        print("4. Exit")
        choice = input("Enter your choice (1-4): ")
        print(choice)

        if choice == "1":
            for rex in passwords:
                print(rex)
        elif choice =="2":
            add_password(passwords)
        elif choice =="3":
            search_password(passwords)
        elif choice =="4":
            break
        else: 
            print("Please enter the valid selection")



        
```

```python
main()
```
