```python
def load_tasks():
    tasks = []

    try:
        with open("todo.txt","r") as file:
            for line in file:
                tasks.append(line.strip())
    except FileNotFoundError:
        pass
    return tasks
```

```python
def save_tasks(tasks):
    with open("todo.txt", "w") as file:
        for task in tasks:
            file.write(task + "\n")

save_tasks(["Buy milk", "Learn Python"])
```

```python
def show_task(tasks):
    if not tasks:
        print("No tasks")
    else:
        print("Your To-Do List:")
        for index,task in enumerate(tasks, start=1):
            print(f"{index}. {task}")
show_task(["Buy milk", "Learn Python"])
```

```python
def add_task(tasks):
    user_task = input("Enter the task for your list: ")
    tasks.append(user_task)
    save_tasks(tasks)
    print("Sucessfully Added")
t = load_tasks()
add_task(t)
show_task(t)
```

```python
def delete_task(tasks):
    show_task(tasks)
    delete_index = input("Enter the task number you would like to delete: ")
    index = int(delete_index)

    if 1<= index <= len(tasks):
        tasks.pop(index - 1)
        save_tasks(tasks)
        print("Sucessfully Deleted")
    else:
        print("Invalid attempt")
    
    
```

```python
def main():
    tasks = load_tasks()

    while True:
        print("\n===== TO-DO LIST MENU =====")
        print("1. View tasks")
        print("2. Add task")
        print("3. Delete task")
        print("4. Exit")

        choice = input("Choose an option (1-4): ")

        if choice == "1":
            show_task(tasks)
        elif choice == "2":
            add_task(tasks)
        elif choice == "3":
            delete_task(tasks)
        elif choice == "4": 
            break
        else: 
            print("Invalid choice")

main()
```
